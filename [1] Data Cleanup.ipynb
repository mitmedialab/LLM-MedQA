# 1) Data Clean Up

```
import pandas as pd
import matplotlib.pyplot as plt
```

### Function - Clean up Datasets

```
def clean_up(survey, dest_directory):
    
    Anonymous = survey.loc[:, "DistributionChannel"]
    Consent = survey.loc[:,"Consent"]
    Attention_1 = survey.loc[:, "ps_0"]
    Attention_2 = survey.loc[:, "ps_1"]
    Final_screen = survey.loc[:, "screener_2"]
    Finished = survey.loc[:, "Finished"]
    Prolific_ID = survey.loc[:, "prolific_id"]
    
    rows_to_remove = []

    # ANONYMOUS
    for anon in range(len(Anonymous)):
        if Anonymous[anon] != "anonymous":
            rows_to_remove.append(anon)
             
    # CONSENT
    for cons in range(len(Consent)):
        if Consent[cons] == "I do not consent":
            if cons not in rows_to_remove:
                rows_to_remove.append(cons)
  
    # ATTENTION 1
    for att1 in range(len(Attention_1)):
        if Attention_1[att1] != "15":
            if att1 not in rows_to_remove:
                rows_to_remove.append(att1)
                              
    # ATTENTION 2    
    for att2 in range(len(Attention_2)):
        if Attention_2[att2] != "Somewhat disagree":
            if att2 not in rows_to_remove:
                rows_to_remove.append(att2)
     
    # FINAL SCREEN  
    for finsc in range(len(Final_screen)):
        if Final_screen[finsc] != "Red,Green":
            if finsc not in rows_to_remove:
                rows_to_remove.append(finsc) 
   
    # FINISHED
    for fin in range(len(Finished)):
        if Finished[fin] == "FALSE":
            if fin not in rows_to_remove:
                rows_to_remove.append(fin) 
    
    survey_2 = survey.copy()

    # REMOVED ROWS
    for row in rows_to_remove:
        
        survey_2 = survey_2.drop(row)

    # PROLIFIC ID
    survey_2.drop_duplicates(subset=['prolific_id'])

    new_index_numbers = []
    for i in range(len(survey_2)):
        new_index_numbers.append(i)
    
    survey_2.insert(0, "index", new_index_numbers)

    survey_2.set_index('index', inplace = True)

    survey_2.to_csv(dest_directory) 

    return("Data all cleaned")
```

### Survey Datasets

```
#Load Dataset - SURVEY 1
Survey_1_DS = pd.read_csv("/work/data/1) raw_data/Edit - SURVEY1 - GTP3 - Perception - Copy_April 13, 2023_07.30.csv")
S1_directory = "/work/data/2) cleaned_data/Cleaned - Survey 1.csv"
clean_up(Survey_1_DS, S1_directory)

#Survey_1 = Survey_1.drop(columns=Survey_1.columns[[0]])
```
